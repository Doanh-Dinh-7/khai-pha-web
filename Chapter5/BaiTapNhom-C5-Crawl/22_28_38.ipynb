```python
import requests
import re
import os
from bs4 import BeautifulSoup 
import html5lib
from urllib.parse import  urljoin, urlsplit
```

```python
url_fileRSSs=["https://dantri.com.vn/rss/giao-duc.rss", "https://dantri.com.vn/rss/giai-tri.rss", "https://dantri.com.vn/rss/du-lich.rss"]
index=1
for url_fileRSS in url_fileRSSs:
    file_store=f'D:/PythonRaymond/LinhTinh/KPW/Chapter5/22_28_38/Data/News{index}'
    print("Processing RSS file:", url_fileRSS)
    
    domain= urlsplit(url_fileRSS).netloc
    stt = 1

    # def url_to_file_name(url):
    #     url = str(url).strip().replace(' ', '_')
    #     return re.sub(r'(?u)[^-\w.]', '', url) + ".txt"

    def save_toTXT(url, stt):
        print("Saving file,...")
        
        html = requests.get(url).text
        soup = BeautifulSoup(html, "html5lib")
        try:
            Title = soup.find("h1", class_="title-page detail")
            Desc = soup.find("h2", class_="singular-sapo")
            Detail = soup.find("div", class_="singular-content")
            Content= Title.text + " " + Desc.text + " " + Detail.text
            
            filename=os.path.join(file_store, f'News{index}{stt}.txt')    

            with open(filename, 'w',encoding='utf-8') as f:
                f.write(Content)
            if stt == 20:
                return
            return stt + 1  # Tăng stt sau mỗi lần lưu
        except Exception as e:
            return stt

                    
    def visit(url, stt):
        print("** Now visiting:", url)       
        stt = save_toTXT(url, stt)
        return stt

    def getURL_from_fileRSS(url_fileRSS):
        xml = requests.get(url_fileRSS).text
        links=re.findall('<link>(.*?)</link>', xml)
        
        links_todo=[]    
        for link_url in links:
            if link_url is None:
                continue      
            
            if urlsplit(link_url).netloc[3:] != domain[3:]:
                continue
            
            if link_url in links_todo:
                continue
            
            links_todo.append(link_url)
            
        return links_todo
    
    links_todo = getURL_from_fileRSS(url_fileRSS)
    
    while links_todo:
        url_to_visit = links_todo.pop()
        stt = visit(url_to_visit, stt)
    index+=1
```
