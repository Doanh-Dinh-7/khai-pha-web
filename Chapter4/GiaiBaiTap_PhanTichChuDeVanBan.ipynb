```python
#ĐỌC FILE
import glob
import os

folder_path = 'data/2/'
# Lấy danh sách tất cả các file .txt trong thư mục
all_files = glob.glob(os.path.join(folder_path, "*.txt"))

# Kết hợp nội dung của tất cả các file
combined_content = ""
for file in all_files:
    with open(file, 'r', encoding='UTF-16') as f:
        combined_content += f.read() + "\n"

# In toàn bộ nội dung kết hợp
print("Nội dung kết hợp của tất cả các file:")
print(combined_content)
```

```python
#TÁCH CÂU
import re

text=combined_content
# Tách câu bằng biểu thức chính quy
sentences = re.split(r'(?<=[.!?]) +', text)

# In danh sách các câu
for i, sentence in enumerate(sentences, start=1):
    sentences[i-1] = re.sub(r'\.', '', sentence)

print(sentences)
```

```python
#XÓA DẤU CÂU
import re

for i, sentence in enumerate(sentences, start=1):
    sentences[i-1] = re.sub(r'[^\w\s.]', '', sentence)
print(sentences)
```

```python
#TÁCH TỪ
# !pip install pyvi
from pyvi import ViTokenizer
for i, sentence in enumerate(sentences, start=1):
   #  sentences[i-1] = re.sub(r"\d+", "", sentence)
    sentences[i-1]=ViTokenizer.tokenize(sentence)

print(sentences)
```

```python
#CHUYỂN CHỮ THƯỜNG
for i, sentence in enumerate(sentences, start=1):
    sentences[i-1] = sentence.lower()

print(sentences)

#XÓA SỐ
for i, sentence in enumerate(sentences, start=1):
    sentences[i-1] = re.sub(r"\d+", "", sentence)
    sentences[i-1] = sentences[i-1].replace("  "," ")
print(sentences)
```

```python
#XÓA STOPWORDS
import os
from pathlib import Path

# text=sentences

notebook_path = Path(os.getcwd()) / 'data/vietnamese-stopwords.txt'
# Đọc file stopwords
with open(notebook_path, "r", encoding="utf-8") as f:
    # Lấy danh sách stopwords
    List_StopWords = [word.strip() for word in f.read().split("\n") if word.strip()]

# Loại bỏ stopwords khỏi văn bản
for i, sentence in enumerate(sentences, start=1):
    sentences[i-1] = re.sub(r"\d+", "", sentence)
    sentences[i-1]=" ".join(word for word in sentence.split() if word not in List_StopWords)

print(sentences)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
import pandas as pd

D=sentences

# Tạo mô hình TF-IDF
tfidf_vectorizer = TfidfVectorizer()
tfidf_matrix = tfidf_vectorizer.fit_transform(D)

tfidf_df = pd.DataFrame(tfidf_matrix.toarray(), columns=tfidf_vectorizer.get_feature_names_out())

# Hiển thị các từ có TF-IDF cao nhất cho mỗi tài liệu
top_n = 3
top_keywords_per_doc = {}

for i, row in enumerate(tfidf_df.iterrows()):
    sorted_row = row[1].sort_values(ascending=False)  # Sắp xếp các từ trong tài liệu theo TF-IDF giảm dần
    top_keywords_per_doc[f"Document {i+1}"] = sorted_row.head(top_n).index.tolist()

# Hiển thị từ khóa TF-IDF cao nhất của từng tài liệu
print("Từ khóa có trọng số TF-IDF cao nhất trong mỗi tài liệu:")
for doc, keywords in top_keywords_per_doc.items():
    print(f"{doc}: {keywords}")

# Hiển thị các từ khóa có TF-IDF cao nhất trong toàn bộ tập dữ liệu
top_keywords_corpus = tfidf_df.max().sort_values(ascending=False).head(10)
print("\nTừ khóa có trọng số TF-IDF cao nhất trong toàn bộ tập dữ liệu:")
print(top_keywords_corpus)
```
