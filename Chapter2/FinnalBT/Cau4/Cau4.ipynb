```python
import os
import re
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
```

```python
# Thiết lập trình duyệt
driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()))

# Đọc các đường link từ file txt
file_path = "Cau2-4.txt"  # Đường dẫn tới file .txt chứa các đường link
with open(file_path, 'r') as file:
    urls = file.readlines()

# Tạo thư mục để lưu các file HTML (nếu chưa có)
output_dir = "HTML"
if not os.path.exists(output_dir):
    os.makedirs(output_dir)

# Duyệt qua từng URL và lưu mã nguồn HTML
for url in urls:
    url = url.strip()  # Loại bỏ ký tự xuống dòng hoặc khoảng trắng thừa
    if not url:
        continue

    # Lấy toàn bộ mã nguồn HTML của trang
    driver.get(url)
    html_content = driver.page_source

    # Tạo tên file từ đường link (loại bỏ các ký tự không hợp lệ)
    file_name = re.sub(r'[^\w\-_\. ]', '_', url) + '.html'

    # Đường dẫn để lưu file trong thư mục đã tạo
    file_path = os.path.join(output_dir, file_name)

    # Lưu mã nguồn HTML vào file
    with open(file_path, 'w', encoding='utf-8') as file:
        file.write(html_content)

    print(f"Đã lưu mã nguồn HTML của {url} vào file {file_path}")

# Đóng trình duyệt
driver.quit()
```
